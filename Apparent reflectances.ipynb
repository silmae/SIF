```python
import holoviews as hv
hv.extension('bokeh')
```

```python
import sif
import numpy as np
import pandas as pd
```

```python
df = sif.read_SIF_data('Tree2_090518_Summary.csv')
df.head()
```

```python
df = sif.compute_sun_pos(df, sif.FOREST)
```

```python
def R_fit(L, E):
    return np.linalg.lstsq(E.reshape(E.size, 1), L.reshape(L.size, 1), rcond=None)
```

```python
def leaf_plot(azi, alt):
    leaf = np.array([[azi, alt]])
    res = sif.apply_cosine(df, leaf)
    p1 = hv.Points(res, kdims=['E_760', 'L_760'])
    p2 = hv.Points(res, kdims=['E_757', 'L_757'])
    p3 = hv.Points(res, kdims=['E_685', 'L_685'])
    p4 = hv.Points(res, kdims=['E_687', 'L_687'])
    return hv.Layout(p1 + p2 + p3 + p4).cols(2)
```

```python
hv.DynamicMap(
    leaf_plot, kdims=['azi', 'alt']
).redim.range(
    azi=(0, 360),
    alt=(0,90)
)
```

```python
def R_plot(azi, alt):
    start_t = '12:00'
    end_t = '13:00'
    leaf = np.array([[azi, alt]])
    res = sif.apply_cosine(df, leaf)
    roi = res.tz_convert('US/Eastern').between_time(start_t, end_t)
    
    wls = [757, 760, 685, 687]
    p = []
    for k,wl in enumerate(wls):
        R = R_fit(L=roi[f'L_{wl}'].values, E=roi[f'E_{wl}'].values)[0].squeeze()
        p.append(
            hv.Points(res, kdims=[f'E_{wl}', f'L_{wl}']) * \
            hv.Points(roi, kdims=[f'E_{wl}', f'L_{wl}']) * \
            hv.Curve(
                [[0, 0], [roi[f'E_{wl}'].max(), R * roi[f'E_{wl}'].max()]], label=f'{np.round(R, decimals=3)}'
                ).options({'Curve':{'color':'orange'}})
            )
            
    return hv.Layout(p[0] + p[1] + p[2] + p[3]).cols(2)

hv.DynamicMap(
    R_plot, kdims=['azi', 'alt'],
).redim.range(
    azi=(0, 360),
    alt=(0,90)
)
```
