```python
import sif
import numpy as np
```

```python
import holoviews as hv
hv.extension('bokeh')
```

```python
df1 = sif.read_SIF_data('Tree2_090518_Summary.csv')
df2 = sif.read_SIF_data('Tree2_090618_Summary.csv')
```

```python
df1 = sif.compute_sun_pos(df1, sif.FOREST).tz_convert('US/Eastern')
df2 = sif.compute_sun_pos(df2, sif.FOREST).tz_convert('US/Eastern')
```

```python
def comparison_plot(azi, alt, start_t, end_t):
    leaf = np.array([[azi, alt]])
    res1 = sif.apply_cosine(df1, leaf)
    res2 = sif.apply_cosine(df2, leaf)
    roi1 = res1.between_time(start_t, end_t)
    roi2 = res2.between_time(start_t, end_t)
    
    wls = [757, 760, 685, 687]
    p = []
    for k,wl in enumerate(wls):
#        R = R_fit(L=roi[f'L_{wl}'].values, E=roi[f'E_{wl}'].values)[0].squeeze()
#        Emin, Emax = roi[f'E_{wl}'].min(), roi[f'E_{wl}'].max()
#        absmax = np.float64(np.where(-Emin > Emax, Emin, Emax))
        p.append(
            #hv.Points(res1, kdims=[f'E_{wl}', f'L_{wl}']) * hv.Points(res2, kdims=[f'E_{wl}', f'L_{wl}'])
            hv.Points(roi1, kdims=[f'E_{wl}', f'L_{wl}']) * hv.Points(roi2, kdims=[f'E_{wl}', f'L_{wl}'])
#            hv.Curve(
#                [[0, 0], [absmax, R * absmax]], label=f'{np.round(R, decimals=3)}'
#                ).options({'Curve':{'color':'orange'}})
            )
            
    return hv.Layout(p[0] + p[1] + p[2] + p[3]).cols(2)

hv.DynamicMap(
    comparison_plot, kdims=['azi', 'alt', 'start_t', 'end_t'],
).redim(
    azi=hv.Dimension('azimuth', default=150, range=(0, 360)),
    alt=hv.Dimension('altitude', default=70, range=(0,90)),
    start_t=hv.Dimension('start_t', label='Start time', default=[t.time() for t in df1.index][0], values=[t.time() for t in df1.index]),
    end_t=hv.Dimension('end_t', label='End time', default=[t.time() for t in df1.index][-1], values=[t.time() for t in df1.index])
)
```
