```python
import holoviews as hv
hv.extension('bokeh')
```

```python
import sif
import numpy as np
import pandas as pd
```

```python
#df = sif.read_SIF_data('Tree2_090418_Summary.csv')
#df = sif.read_SIF_data('Tree2_090518_Summary.csv')
df = sif.read_SIF_data('Tree2_090618_Summary.csv')
df.head()
```

```python
df = sif.compute_sun_pos(df, sif.FOREST)
```

```python
def R_fit(L, E):
    return np.linalg.lstsq(E.reshape(E.size, 1), L.reshape(L.size, 1), rcond=None)
```

```python
def R_plot(azi, alt):
    start_t = '13:00'
    end_t = '14:00'
    leaf = np.array([[azi, alt]])
    res = sif.apply_cosine(df, leaf)
    roi = res.tz_convert('US/Eastern').between_time(start_t, end_t)
    
    wls = [757, 760, 685, 687]
    p = []
    for k,wl in enumerate(wls):
        R = R_fit(L=roi[f'L_{wl}'].values, E=roi[f'E_{wl}'].values)[0].squeeze()
        Emin, Emax = roi[f'E_{wl}'].min(), roi[f'E_{wl}'].max()
        absmax = np.float64(np.where(-Emin > Emax, Emin, Emax))
        p.append(
            hv.Points(res, kdims=[f'E_{wl}', f'L_{wl}']) * \
            hv.Points(roi, kdims=[f'E_{wl}', f'L_{wl}']) * \
            hv.Curve(
                [[0, 0], [absmax, R * absmax]], label=f'{np.round(R, decimals=3)}'
                ).options({'Curve':{'color':'orange'}})
            )
            
    return hv.Layout(p[0] + p[1] + p[2] + p[3]).cols(2)

hv.DynamicMap(
    R_plot, kdims=['azi', 'alt'],
).redim(
    azi=hv.Dimension('azimuth', default=150, range=(0, 360)),
    alt=hv.Dimension('altitude', default=70, range=(0,90))
)
```
